```python
# Keep only beers from sysbo.
mask = sysbo_df['group'] == 'Beer'
sysbo_df = sysbo_df[mask]
sysbo_df.head()
```

```python
mask = sysbo_df['group']
mergedStuff = sysbo_df.merge(beeradv_df, left_on=['abv'], right_on=['beer_abv'], how='inner')
```

```python
mergedStuff.info()
```

```python
from textdistance import levenshtein
from time import time

start = time()
print(mergedStuff.shape)

mini = mergedStuff[:100000]

mini['lev'] = mini.apply(lambda x: levenshtein.distance(x['name'],  x['beer_name']), axis=1)
print(time()-start)
```

```python
mini[['number', 'name', 'group', 'abv', 'beer_name', 'beer_id', 'lev']].sort_values(by='lev', ascending=True)[:40]
```
